# Top ten areas with the lowest median salary

```python
import pandas as pd

# Load the CSV data
url = "https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv"
data = pd.read_csv(url)

# Find the bottom ten areas with the smallest median salary
bottom_ten_areas = data.nsmallest(10, 'median_salary')

# Display the data for the bottom ten areas
display(bottom_ten_areas)
```
